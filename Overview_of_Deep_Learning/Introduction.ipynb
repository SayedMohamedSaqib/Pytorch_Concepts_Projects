```python
### Pytorch Modules
import torch 
```

```python
# Instantiating a tensor

points = torch.tensor([[1.0, 4.0], 
                       [2.0, 1.0], 
                       [3.0, 5.0]])
```

```python
points[0]
```

```python
# We can also check the shape ofthe tensor using this

points.shape
```

```python
"""  tensors are implemented as views over a one-dimensional array
of numerical data stored in contiguous chunks ofmemory. These arrays are
called storage instances. """

points.storage()
```

```python
points.size()
```

```python
# The offset here represents the index ofthe first element ofthe tensor in the storage array.

points.storage_offset() 
```

```python
points[1].storage_offset() 
```

```python
#  stride contains, for each dimension, the number of elements to be skipped in order to access the next element ofthe tensor.

points.stride() 
```

```python
# we specify a certain data type to be used for a PyTorch tensor is as follows:

points = torch.tensor([[1.0, 2.0], [3.0, 4.0]], dtype = torch.float32)
```

```python
# Besides the data type, tensors in PyTorch also need a device specification where they will be stored. A device can be specified as instantiation:
points = torch.tensor(
    [[1.0, 2.0], [3.0, 4.0]],
    dtype=torch.float32,
    device="cpu"
)
```
