```python
"""Under the utils.data module, Torch provides its own dataset and
DataLoader classes, which are extremely handy due to their abstract and
flexible implementations. """
from torch.utils.data import TensorDataset, DataLoader

# reproducibility
torch.manual_seed(0)
x_train = torch.randn(100, 256)


y_train = (x_train.sum(dim=1) > 0).long()
train_dataset = TensorDataset(x_train, y_train)
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)


for xb, yb in train_dataloader:
    print(xb.shape, yb.shape)
    break
```

```python
import torch
import torch.nn as nn

# simple model: 256 → 64 → 4
class SimpleNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(256, 64),
            nn.ReLU(),
            nn.Linear(64, 4)
        )
    
    def forward(self, x):
        return self.net(x)

# instantiate model
model = SimpleNet()

# use it with your dataloader
for x_batch, y_batch in train_dataloader:
    pred = model(x_batch)
    print(pred.shape)  # should be (batch_size, 4)
    break
```
